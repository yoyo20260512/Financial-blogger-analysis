```python
# ===== 一站式：加载 + 测试（不依赖任何其他单元格）=====
import torch, gc, warnings
warnings.filterwarnings("ignore")
gc.collect()
torch.mps.empty_cache()

from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

print("🚀 正在加载 Qwen2.5 + LoRA ...")
device = "mps"
BASE = "/Users/jinqianyu/Documents/个人资料/大数据/财经博主/models/Qwen2.5-1.5B-Instruct"
LORA = "/Users/jinqianyu/Documents/个人资料/大数据/财经博主/outputs/qwen_lora_finetuned"

tokenizer = AutoTokenizer.from_pretrained(BASE, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token

base = AutoModelForCausalLM.from_pretrained(BASE, trust_remote_code=True, torch_dtype=torch.float16, device_map=device)
model = PeftModel.from_pretrained(base, LORA)
model.eval()
print("✅ 加载完成！\n")

# 测试生成
def ask(q):
    text = tokenizer.apply_chat_template([{"role":"user","content":q}], tokenize=False, add_generation_prompt=True)
    out = model.generate(**tokenizer(text, return_tensors="pt").to(device), max_new_tokens=200, do_sample=False)
    return tokenizer.decode(out[0][len(tokenizer(text,return_tensors='pt')['input_ids'][0]):], skip_special_tokens=True)

for q in [
    "什么是LPR？对房贷有什么影响？",
    "什么是市盈率（PE）？如何使用？",
    "A股、港股和美股的主要区别是什么？",
]:
    print(f"📌 问: {q}")
    print(f"💡 答: {ask(q)}")
    print("="*60)
```
